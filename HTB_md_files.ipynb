```python
import os
import re
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
from pathlib import Path

# --- Configuration ---
BASE_URL = "https://0xdf.gitlab.io"
TREATED_LINKS_FILE = "treated_links2.txt"
LINKS_TO_TREAT = 15

# Raw excluded titles (includes HTB and variations)
RAW_EXCLUDED_TITLES = {
    'reel HTB writeup', 'HTB cascade', 'Aero HTB', 'Celestial HTB', 'MONTEVERDE HTB',
    'lightweight', 'coneal HTB', 'Legacy HTB', 'Optimum htb', 'blue HTB', 'Ready HTB',
    'Devel HTB', 'HTB Ophiuchi', 'Bashed HTB', 'Bounty HTB', 'Schooled HTB', 'Kiba TRH',
    'Bank HTB', 'atom HTB', 'Horizontall HTB', 'Inception HTB', 'intelligence HTB',
    'Jarvis HTB', 'jeeves HTB', 'Poison HTB', 'hackback HTB', 'Player HTB', 'Ghoul HTB',
    'ypuffy HTB', 'zipper HTB', 'Json HTB', 'Caption HTB', 'permx HTB', 'quick HTB',
    'APT HTB', 'htb-busqueda', 'Monteverde HTB', 'Arkham:HTB', 'vessel HTB', 'Spider HTB',
    'Pikaboo', 'Epsilon HTB', 'Hancliffe HTB', 'Overflow HTB', 'jerry HTB', 'Topology HTB',
    'Bagel HTB', 'forget HTB', 'Awkward HTB', 'Shared HTB', 'Noter HTB', 'talkative HTB',
    'router space', 'artcorp HTB', 'Antique HTB', 'MIrai HTB', 'Backdoor HTB', 'timing',
    'UpDown HTB', 'Faculty HTB', 'Previse HTB', 'Precious HTB', 'Bounty hunter HTB',
    'timelapse HTB', 'RedPanda HTB', 'RainyDay HTB', 'shibboleth HTB', 'Active HTB',
    'Bizness HTB', 'sandworm HTB', 'pentest-report_whiteout', 'investigation HTB',
    'Acute HTB', 'Trick HTB', 'netmon HTB', 'Explore HTB', 'Static HTB', 'Pandora HTB ',
    'OpenSource HTB', 'MetaTwo HTB', 'PhotoBomb HTB'
}

# Normalize excluded slugs (extract only lowercase names without "HTB", extra spaces, etc.)
def clean_name(title):
    return re.sub(r'[^a-z0-9]+', '', title.lower().replace("htb", "").strip())

EXCLUDED_SLUGS = set(clean_name(title) for title in RAW_EXCLUDED_TITLES)

# Load already treated links
if not os.path.exists(TREATED_LINKS_FILE):
    open(TREATED_LINKS_FILE, 'w').close()

with open(TREATED_LINKS_FILE, 'r') as f:
    treated_links = set(line.strip() for line in f)

# Fetch page and extract new article links
response = requests.get(BASE_URL)
response.raise_for_status()
soup = BeautifulSoup(response.text, "html.parser")

post_links = []
for a in soup.select("a[href^='/20']"):
    full_url = urljoin(BASE_URL, a["href"])
    path = urlparse(full_url).path
    filename = os.path.basename(path)

    # Skip if filename has multiple dashes (often index pages or multi-topic)
    if filename.count('-') >= 2:
        continue

    # Extract the machine name from the filename: e.g., "htb-bizness.html" → "bizness"
    name_slug = clean_name(filename.replace("htb-", "").replace(".html", ""))

    # Check exclusion
    if name_slug in EXCLUDED_SLUGS:
        continue

    if full_url not in treated_links and full_url not in post_links:
        post_links.append(full_url)

# Take top N
selected_links = post_links[:LINKS_TO_TREAT]

print(f"📄 Liens sélectionnés ({len(selected_links)}) :")
for link in selected_links:
    print(" -", link)
```

```python
# Traitement de chaque article sélectionné
import requests
from bs4 import BeautifulSoup
from bs4.element import Tag
from urllib.parse import urljoin, urlparse
import os
from pathlib import Path
import re
from markdownify import markdownify as md

for url in selected_links:
    print(f"\n🔗 Traitement de : {url}")
    try:
        res = requests.get(url)
        res.raise_for_status()
        soup = BeautifulSoup(res.text, "html.parser")

        postbody = soup.find("div", id="postBody")
        start_tag = postbody.find("h2", id="recon")
        end_tag = postbody.find(id="beyond-root")

        content = []
        reached = False
        for child in postbody.children:
          if isinstance(child, Tag):
             if child == start_tag:
                reached = True
             if reached:
               if child == end_tag:
                 break  # stop when reaching the tag with id="beyond-root"
               content.append(child)


        new_soup = BeautifulSoup("".join(str(tag) for tag in content), "html.parser")

        os.makedirs("images", exist_ok=True)
        for img in new_soup.find_all("img"):
            img_url = img.get("src")
            if not img_url:
                continue
            full_img_url = urljoin(BASE_URL, img_url)
            try:
                img_data = requests.get(full_img_url).content
                filename = os.path.basename(urlparse(img_url).path)
                filepath = os.path.join("images", filename)
                with open(filepath, "wb") as f:
                    f.write(img_data)
                img["src"] = filepath
            except Exception as e:
                print(f"⚠️ Erreur image : {full_img_url} -> {e}")

        for tag in new_soup.find_all(["script", "style"]):
            tag.decompose()

        markdown_text = md(str(new_soup))

        file_base = url.split("/")[-1].replace(".html", "")
        md_path = Path(f"{file_base}.md")
        modified_md_path = Path(f"modified_{file_base}.md")

        with open(md_path, "w", encoding="utf-8") as f:
            f.write(markdown_text)

        print("✅ Markdown généré à partir de <h2 id='recon'>")

        with open(md_path, "r", encoding="utf-8") as f:
            md_content = f.read()

        # Nettoyage des balises "expand"
        expand_pattern = r'\[!\[expand\]\(images/expand\.png\)\]\(javascript:void\(0\).*?\)'
        md_content = re.sub(expand_pattern, '', md_content)

        # Statistiques des images
        img_pattern = r"!\[.*?\]\((images/.*?)\)"
        matches = re.findall(img_pattern, md_content)
        print(f"🖼️ Images restantes à traiter : {len(matches)}")

        with open(modified_md_path, "w", encoding="utf-8") as f:
            f.write(md_content)

        print(f"✅ Fichier final : {modified_md_path.name}")

        # Mise à jour du fichier de suivi
        with open(TREATED_LINKS_FILE, "a") as f:
            f.write(url + "\n")

    except Exception as e:
        print(f"❌ Erreur : {url} -> {e}")
```
